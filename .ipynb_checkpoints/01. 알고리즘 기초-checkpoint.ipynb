### 세 정수의 최댓값 구하기

````python
# 실습 1-1.입력받은 세 정수의 최댓값 구하기

a = int(input('정수 a 값을 입력하시오 : '))  # input()함수로 입력값 받을 수 있음. 입력받은 데이터의 자료형은 문자열이므로 int 통해 형변환.
b = int(input('정수 b 값을 입력하시오 : '))
c = int(input('정수 c 값을 입력하시오 : '))

max_num = a

if b > max_num:
    max_num = b
if c > max_num:
    max_num = c;

print('최댓값은 '+ max_num +'입니다')  # str(문자열)와 int는 + 로 연결할 수 없다.
````

````python
# 실습 1-1. 입력받은 세 정수의 최댓값 구하기

a = int(input('정수 a 값을 입력하시오 : '))
b = int(input('정수 b 값을 입력하시오 : '))
c = int(input('정수 c 값을 입력하시오 : '))

max_num = a

if b > max_num:
    max_num = b
if c > max_num:
    max_num = c;

print(f'최댓값은 {max_num}입니다.')  # print 함수 내에서 문자열과 정수형 함수 연결해 함께 출력하려면 포맷 문자열 리터럴 사용.
````

````python
# 보충 수업 1-1. 문자열과 숫자 입력받기

print('이름을 입력하시오 ')
name = input()  # 왜 input()을 통해 입력된 데이터는 출력된다. 파이썬은 개행이 기본이라 위 print문 아랫줄에 출력됨.
print('안녕하세요 '+ name + '님!')  # print 함수 내에서 str형은 + 로 연결해 출력 가능.
print(f'안녕하세요 {name}님!')  # 포맷 문자열 리터럴 사용해도 결과는 같다.
````

````python
# 실습 1-2. 세 정수의 최댓값 구하기

def max_num(a,b,c):  # 매개변수 a,b,c에 숫자 입력받아 최대값 반환.
    maximum = a
    if b > maximum:
        maximum = b
    if c > maximum:
        maximum = c
    return maximum  # 함수 내부에서 처리한 값 반환할 땐 return문 사용.

print(f'max_num(3,2,2) = {max_num(3,2,2)}')
print('max_num(1,2,3) = ' + max_num(1,2,3))  # print 함수 내에서 int와 str +로 연결 불가. 포맷 문자열 리터럴 사용.
````

**보충 수업 1-2. 복합문의 구조** <br>
복합문이란 하나 또는 그 이상의 절로 구성되는 것. 헤더(header)와 스위트(suite)로 구성된다.

```python
if a < b:   # 헤더. if 나 while과 같은 키워드로 시작해 콜론으로 끝나는 부분
    min_num = a  # 스위트. 헤더와 한 세트로 따라다니는 실행문
```

**보충수업 1-4. 복합문 작성 규칙** <br>
스위트가 단순문이라면 => 헤더와 같은 행에 위치 가능.

```python
if a < b : min_num = a  # 단순문 1개
if a < b : min_num = a; max_num = b  # 단순문 2개 (마지막 단순문에 세미콜론 붙이는건 자유! 아래 코드 참고.)
if a < b : min_num = a; max_num = b; <br>
```


````python
# 실습 1C-2. 세 정수의 대소 관계와 중앙값 구하기
# 중앙값(median)이란 주어진 값을 크기 순서대로 정렬했을 때 가장 중앙에 위치하는 값. 중앙값 != 평균값(mean)

def med_num(a,b,c):
    if a >= b:  # 스위트가 if문으로 이루어진 복합문이므로 다음 행에 작성.
        if b >= c:  # a>=b>= c
            return b
        elif a <=c:  # c>=a>=b
            return a
        else:  # b>=c가 아니고, a<=c가 아닌 경우는 c<b and a>c이므로 a>c>b
            return c
    elif a > c:  # a<b를 만족하고, a>c인 경우이므로 b>a>c
        return a
    elif b > c:  # a<b and a<c를 만족하고, b>c인 경우이므로 b>c>a
        return c
    else:  # a<b and a<=c and b<=c를 만족하는 경우는 a<b<=c
        return b

a = int(input('정수 a의 값을 입력하시오 : '))
b = int(input('정수 b의 값을 입력하시오 : '))
c = int(input('정수 c의 값을 입력하시오 : '))
print(f'세 정수의 중앙값은 {med_num(a,b,c)} 입니다.')
````

````python
# 실습 1C-2. 세 정수의 대소 관계와 중앙값 구하기 2
# 해당 코드는 위 코드보다 비효율적. -> if문의 조건식이 거짓이면 elif문이 이 판단을 수행할 필요가 없기 때문.

def med_num(a,b,c):
    if (b >=a and c <= a) or (b <= a and c >= a):  # 파이썬에선 &&, || 대신 and와 or 연산자 사용.
        return a
    elif (a > b and c < b) or (a < b and c > b):
        return b
    return c

a = int(input('정수 a의 값을 입력하시오 : '))
b = int(input('정수 b의 값을 입력하시오 : '))
c = int(input('정수 c의 값을 입력하시오 : '))
print(f'세 정수의 중앙값은 {med_num(a,b,c)} 입니다.')
````

### 조건문과 분기
**분기(branching)는 프로그램의 실행 흐름을 다른 곳으로 변경하는 명령.**

````python
# 실습 1-4. 3개로 분기하는 조건문

n = int(input('정수를 입력하시오 : '))

if n == 1: print('A')
elif n == 2: print('B')
else: print('C')
# n이 1,2,그 외의 값일 떄의 경우가 있으므로 3개로 분기하는 조건문이다.
````

````python
# 실습 1-4. 4개로 분기하는 조건문
n = int(input('정수를 입력하시오 : '))

if n == 1: print('A')
elif n == 2: print('B')
elif n ==3 : print('C')
# n이 1,2,3일 때와 그 외의 값일 때(else: pass)의 경우가 있으므로 4개로 분기하는 조건문이다.
````

**보충수업 1-6. 연산자(operator)와 피연산자(operand)** <br>
a > b 에서 연산자는 >, 피연산자는 a와 b이다.<br>
<br>
**연산자의 종류 (피연산자의 개수에 따라 분류됨.)** <br>
- 단항 연산자 : ex) -a
- 이항 연산자 : ex) a < b
- 삼항 연산자 : 파이썬의 유일한 삼항 연산자는 if~else문. ex) a if b else c
